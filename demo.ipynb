# Easy Task

```python
ssh_R18, sst_R18, ssh_R18_norm, sst_R18_norm, ssh_R72, sst_R72 = loadDataR18( upscale_factor )
```

```python
# SRCNN
srcnn, callbacks, epochs, checkpoint_filepath = compileModel(100)
batch_size = 16
upsampled_ssh_R72 = np.array(
            [bicubicInterpolation(img, ssh_R18[0].shape[0]) for img in ssh_R72]
            )

bicubic_predictions, bicubic_rmse, bicubic_psnr = bicubicMetrics( ssh_R72[366:], ssh_R18[366:] )

# Training ignored
# Uncomment to train the model with your own parameters
# trainModel( srcnn, batch_size, upsampled_ssh_R72, ssh_R18_norm, callbacks, epochs, 1)

srcnn.load_weights(checkpoint_filepath)
save_path = "./Save/exp1_srcnn_plot"
srcnn_predictions, srcnn_rmse, srcnn_psnr = srcnnMetrics( srcnn, upsampled_ssh_R72[366:], ssh_R18[366:])


# Plot Results
print("Average bicubic RMSE (m): ", np.mean(bicubic_rmse))
print("Average bicubic PSNR: ", np.mean(bicubic_psnr))
print("Average srcnn RMSE (m): ", np.mean(srcnn_rmse))
print("Average srcnn PSNR: ", np.mean(srcnn_psnr))

inputSSH = np.array(
            [downsample(img, 4) for img in ssh_R18]
            )
for i in range(3):
    plt.figure()
    plt.imshow(inputSSH[366+i*10], label="Input", cmap="rainbow")

for i in range(3):
    plot_line([bicubic_predictions[i*10], srcnn_predictions[i*10], ssh_R18[366+i*10]], ["Bicubic", "SRCNN", "Ground Truth"], "rainbow",save_path+str(i), center_colormap=False)
    save_path = "./Save/exp1_srcnn_em"

for i in range(0,30,10):
    plot_line(
        [bicubic_predictions[i]-ssh_R18[366+i], srcnn_predictions[i]-ssh_R18[366+i]], 
        ["Bicubic EM", "SRCNN EM"], 
        "seismic",
        save_path+str(i), 
        label="error",
        center_colormap=True
    )
```

```python
# DIPCI

dipci, callbacks, epochs, checkpoint_filepath = compile_model(100, 0.001, 1)

# Training ignored
# Uncomment to train the model with your own parameters
# train(dipci, ssh_R72, sst_R72, ssh_R18_norm, callbacks, epochs)

dipci.load_weights(checkpoint_filepath)
save_path = "./exp1_dipci_plot"
bicubic_predictions, bicubic_rmse, bicubic_psnr = bicubicMetrics( ssh_R72[366:], ssh_R18[366:] )
dipci_predictions, dipci_rmse, dipci_psnr = dipciMetrics( dipci, ssh_R72[366:], sst_R72[366:], ssh_R18[366:])

print("Average bicubic RMSE (m): ", np.mean(bicubic_rmse))
print("Average bicubic PSNR: ", np.mean(bicubic_psnr))
print("Average dipci RMSE (m): ", np.mean(dipci_rmse))
print("Average dipci PSNR: ", np.mean(dipci_psnr))

for i in range(3):
    plot_line(
        [bicubic_predictions[i*10], 
         dipci_predictions[i*10], 
         ssh_R18[366+i*10]], 
        ["Bicubic", "Ours", "Ground Truth"], 
        "rainbow", 
        save_path+str(i), 
        center_colormap=False
    )
save_path = "./exp1_dipci_em"
for i in range(0,30,10):
    plot_line(
        [bicubic_predictions[i]-ssh_R18[366+i], dipci_predictions[i]-ssh_R18[366+i]], 
        ["Bicubic EM", "DIPCI EM"], 
        "seismic",
        save_path+str(i), 
        label="error",
        center_colormap=True
    )
```
